### Extracting the comments using YouTube API

```python
def comment_results(VIDEO_ID):
    request = youtube.commentThreads().list(
        part="snippet",
        videoId = VIDEO_ID,
        maxResults = 10000,
        order = 'relevance'
    )
    response = request.execute()
    return response
```

```python
#import stopwords

import nltk
from nltk.corpus import stopwords

# import TextBlob - used for finding out the polarity and subjectivity
# import Word - for lemmatization

from textblob import Word, TextBlob

# import regular expression to remove the special characters

import re
```

### Steps for preprocessing the data:
Remove the special characters <br>
Convert the entire data to upper or lower case <br>
Split the data <br>
Remove stopwords <br>
Lemmatization <br>
Join and build the model <br>

```python
def remove_special_characters(original_text):
    pattern = r'[^a-zA-Z0-9\s]'
#     pattern = r'[^\w\s]'
    preprocessed_text = re.sub(pattern,' ',original_text)
    return preprocessed_text

def convert_lowercase(original_text):
    return original_text.lower()

def split_data(original_text):
    return original_text.split()

def remove_stopwords(original_data):                                             # takes a list of split words as input
    stop_words = stopwords.words('english')                                        # list of stopwords
    revised_data = [word for word in original_data if word not in stop_words]    
    return revised_data

def lemmatization(original_data):
    new_data = [Word(word).lemmatize() for word in original_data]
    return new_data

def join_data(list_of_words):
    return ' '.join(list_of_words)
```

Now we put all of these individual functions into one single function:

```python
def preprocess_comment(original_text):
    return join_data(lemmatization(remove_stopwords(split_data(convert_lowercase(remove_special_characters(original_text))))))
```

```python
df['Processed_comment'] = df['Comment'].apply(lambda x: preprocess_comment(x))
```

<b>Polarity </b>: represents the degree of positivity or negativity. It is typically measured on a numeric scale ranging from -1 to 1, where -1 represents the most negative sentiment, 0 represents a neutral sentiment, and 1 represents the most positive sentiment.<br>
<b>Subjectivity</b> : refers to the degree of personal opinion, bias, or emotion present in a text, ranging from 0 to 1, where 0 indicates a highly objective or factual text, and 1 indicates a highly subjective or opinionated text.  

```python
df['polarity'] = df['Processed_comment'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['Processed_comment'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

### General consensus among the comments

```python
def overall_results():
    if(df['polarity'].mean()>=0):
        return 'The general consensus among the comments leans towards a positive tone'
    else:
        return 'The general consensus among the comments leans towards a negative tone'
```

```python
def word_cloud():
    text_data = ' '.join(df['Comment'])

    # Generate word cloud
    wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(text_data)

    # Display the word cloud
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

### Views, Likes and Comments

```python
def get_stats(VIDEO_ID):
    request = youtube.videos().list(
        part = "statistics",
        id = VIDEO_ID
    )
    response = request.execute()
    return response
```
